```python
url='https://www.thisisinsider.com/the-20-most-popular-travel-destinations-in-the-world-2016-9'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('h2', class_="slide-title-text")
c_results= results[0].get_text(strip=True).replace('. ', ', ').split(",")
c_results
```

```python
results
```

```python
r = []
```

```python
for result in results:
    r.append(result.get_text(strip=True).replace('. ', ', ').split(","))
    
r
```

```python
data = pd.DataFrame(r)
```

```python
data.loc[8]="12","Barcelona", "Spain"
data.loc[7]="13","Amsterdam", "Netherlands"
data.loc[14]="6","Singapore", "Singapore"
data.loc[15]="5","New York", "United States of America"
```

```python
data.columns= ["rank", "city", "county"]
data["source"]= "Insider_Travel"
```

```python
data[["rank"]] = data[["rank"]].apply(pd.to_numeric)
data= data.sort_values(by="rank", ascending=True)
```

```python
data_t= data.transpose()
data_t
data_j = data_t.to_json()
```

```python
json_data = json.loads(data_j)
type(json_data)
```
