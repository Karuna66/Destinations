```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
import numpy as np
import pandas as pd
```

```python
#executable_path = {'executable_path': 'chromedriver'}

executable_path = {'executable_path':'C:/Users/gairo/Downloads/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# grabbed the states
url='https://alphabetizer.flap.tv/lists/list-of-states-in-alphabetical-order.php'
browser.visit(url)
html=browser.html
soup=bs(html,'lxml')
us_states=[]
results=soup.find_all('li')
for x in range(len(results)):
    us_states.append(results[x].get_text())
```

```python
#cleaned up data
url='https://www.tripadvisor.com/TravelersChoice-Destinations-cTop-g1'

browser.visit(url)
html=browser.html
soup=bs(html,"html.parser")
rating=[]
city=[]
country=[]
id=[]
results=soup.find_all('div',class_='winnerLayer')
for result in results:
    rating.append(result.div.span.get_text())
    place=result.find('div',class_='winnerName').div.a.get_text()
    if len(place.split(',')) ==2:
        city.append(place.split(',')[0])
        country.append(place.split(',')[1])
    else:
        city.append(np.nan)
        country.append(place.split(',')[0])

for x in range(len(country)):
    if (country[x].strip() in us_states):
        country[x]='United States'
        if city[x]==np.nan:
            city[x]=country[x] 
    
trip_advisor_df=pd.DataFrame(dict(rank=rating,city=city,country=country,source="trip_advisor"))
```

```python
trip_advisor_df.set_index('country',inplace=True)
#travel_dest,items
```

```python
#cleaned up data
url="https://www.travelandleisure.com/trip-ideas/best-places-to-travel-in-2018"
browser.visit(url)
html=browser.html
soup=bs(html,'lxml')
```

```python

country=[]
city=[]
counter=0
results=soup.find_all('h3',class_='glide-slide-title')
check_for_doubles=""
for result in results:
    response=result.get_text()
    
    if check_for_doubles != response:
        if 'The 50 Best Places' not in response:

            response_split=response.strip().split(",")

            if len(response_split)==2:

                # two places added together with 'and' in country
                first_part=response_split[0].strip()
                second_part=response_split[1].strip()

                if ' and' in second_part:
                    
                    second_part=second_part.split(' and')[0]
                    
                country.append(second_part)   
                city.append(first_part)
            else:
                country.append(response_split[0].strip())
                city.append(np.nan)
                
        
        check_for_doubles=response
for x in range(len(country)):
    if (country[x] in us_states):
        country[x]='United States'
        if city[x]==np.nan:
            city[x]=country[x]
       
        
rating=[x for x in range(1,len(country)+1)]
travel_land_leisure=pd.DataFrame(dict(rark=rating,city=city,country=country,source="travel and leisure"))
travel_land_leisure
```

```python
# cleaned up
url='https://www.nytimes.com/interactive/2019/travel/places-to-visit.html'
browser.visit(url)
html=browser.html
soup=bs(html,'lxml')
```

```python
results=soup.find_all('div',class_='g-header')
country=[]
city=[]
for result in results:
    response_country=result.div.h2.get_text()
    response_country_2=result.div.h3.get_text()
    if response_country_2 !="":
        city.append(response_country)
        country.append(response_country_2)
    else:
        city.append(np.nan)
        country.append(response_country)

for x in range(len(country)):
    val= country[x]
    if  "." in val or (val in us_states) or (val=='New York City'):
        
        country[x]='United States'
        if city[x]=="":
            city[x]=val
    else:
        if "/" in val:
            country[x]='Denmark'
        elif "," in val:
            country[x]=country[x].split(", ")[1]
rating=[x for x in range(1,len(country)+1)]    
new_york_times_df=pd.DataFrame(dict(rank=rating,city=city,country=country,souce="nytimes"))
new_york_times_df
```
